part 2 - visualization of data using matplotib , pyplots packages


Generate Scatterplot




lets support we collected ages(in years) , height(in cm) , weight(in kg) information from few patients visting a clinic

```python
people = ["ashu","ram","nandu","alok","varsha","milky","vidhu"]

age = [21,12,32,45,37,18,28]
weight = [55,35,77,40,65,68,45]
height = [150,155,160,170,152,175,156]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

Bar plot

```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=16,
          fontstyle='italic')


plt.bar(x=people, height=weight, width=0.1,
        color='red',edgecolor='k',alpha=0.1)


plt.xlabel("People",fontsize=10)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=10)
plt.show()
```

Generate Histogram

```python
import numpy as np
plt.figure(figsize=(5,6))

plt.hist(weight,color='pink',edgecolor='k', alpha=0.70,bins=5)
plt.title("Histogram of patient weight",fontsize=14)
plt.xlabel("Weight in kgs",fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
